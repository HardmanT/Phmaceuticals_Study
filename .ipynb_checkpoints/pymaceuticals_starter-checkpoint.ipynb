### Analysis

- After taking a closer look at Mouse I509 on Capomulin, it looks like the greatest decrease of the tumor volume was between days 20-25 for 4mm3 and between days 30-35 for 4.5mm3. Getting more insight at what went on during that time will be helpful. 
- There is not much correlation between the average tumor volume to the mouse weight at Correlation Coefficient: 0.84
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_mouse_results = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"] )
# Display the data table for preview
study_mouse_results.head()
```

```python
# Checking the number of mice.
mouse_count = study_mouse_results["Mouse ID"].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_rows = study_mouse_results[study_mouse_results.duplicated(subset=["Mouse ID", "Timepoint"])]
print(duplicate_rows)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
c_study_mouse_results = study_mouse_results[study_mouse_results['Mouse ID'] != 'g989']
c_study_mouse_results.head()
```

```python
# Checking the number of mice in the clean DataFrame.
# Checking the number of mice.
mouse_count_clean = c_study_mouse_results["Mouse ID"].nunique()
mouse_count_clean
```

## Summary Statistics

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
summary_statistics_advanced = c_study_mouse_results.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem']).rename(
    columns={'mean': 'Mean', 'median': 'Median', 'var': 'Variance', 'std': 'Std Deviation', 'sem': 'SEM'}
)
# Using the aggregation method, produce the same summary statistics in a single line
summary_statistics_advanced
```

## Quartiles, Outliers and Boxplots

```python
c_study_mouse_results.loc[c_study_mouse_results['Mouse ID'] == "g570",:]
# last_timepoints_df.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens: 
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
target_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
four_regimens = c_study_mouse_results.loc[c_study_mouse_results['Drug Regimen'].isin(target_regimens),:]
maxtime = four_regimens.groupby("Mouse ID")["Timepoint"].max()
maxtime_df = pd.DataFrame(maxtime)
maxtime_df
merge_df = pd.merge(four_regimens, maxtime_df, on="Mouse ID")
merge_df2 = merge_df.loc[merge_df["Timepoint_x"] == merge_df["Timepoint_y"], :]
merge_df2
final_tumor = merge_df2.groupby("Mouse ID")["Tumor Volume (mm3)"].min()
final_tumor_df = pd.DataFrame(final_tumor)
final_tumor_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers.    
    # Locate the rows which contain mice on each drug and get the tumor volumes

    # add subset 
    
    # Determine outliers using upper and lower bounds
treatments_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

tumor_vol_data = []
final_tumor_volume = c_study_mouse_results

for treatment in treatments_list:
    subset_data = final_tumor_volume.loc[final_tumor_volume['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    tumor_vol_data.append(subset_data)
    quartiles = subset_data.quantile([0.25, 0.5, 0.75])
    lower_quartile = quartiles[0.25]
    upper_quartile = quartiles[0.75]
    iqr = upper_quartile - lower_quartile
    lower_bound = lower_quartile - 1.5 * iqr
    upper_bound = upper_quartile + 1.5 * iqr
    print(f"{treatment}:")
    print(f"  - Lower Quartile: {lower_quartile}")
    print(f"  - Upper Quartile: {upper_quartile}")
    print(f"  - IQR: {iqr}")
    print(f"  - Lower Bound: {lower_bound}")
    print(f"  - Upper Bound: {upper_bound}")
    potential_outliers = subset_data[(subset_data < lower_bound) | (subset_data > upper_bound)]
    if not potential_outliers.empty:
        print(f"  - Potential Outliers: {potential_outliers}")
    else:
        print("  - No potential outliers")
    print("\n")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
treatments_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

tumor_vol_data = []

for treatment in treatments_list:
    subset_data = merge_df2.loc[merge_df2['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    tumor_vol_data.append(subset_data)

plt.boxplot(tumor_vol_data, labels=treatments_list, flierprops=dict(marker='o', markerfacecolor='red', markersize=8))

# Labels and titles
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.title('Distribution of Tumor Volume for Each Treatment Group')

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin_random = c_study_mouse_results.loc[c_study_mouse_results["Drug Regimen"] == "Capomulin",:]
capomulin = Capomulin_random.loc[Capomulin_random["Mouse ID"] == "l509",:]

timepoints = capomulin['Timepoint']
tumor_volume = capomulin['Tumor Volume (mm3)']

plt.plot(timepoints, tumor_volume, marker='v', color='purple')

plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Capomulin treatment of Mouse l509')

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_data = c_study_mouse_results[c_study_mouse_results['Drug Regimen'] == 'Capomulin']

average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weight = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

plt.scatter(mouse_weight, average_tumor_volume, marker='o', facecolors='blue', edgecolors='black', alpha=0.75)

plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
from scipy.stats import linregress

correlation_coefficient = st.pearsonr(mouse_weight, average_tumor_volume)[0]

(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight, average_tumor_volume)

regress_values = mouse_weight * slope + intercept

plt.scatter(mouse_weight, average_tumor_volume, marker='s', facecolors='red', edgecolors='black', alpha=0.75)
plt.plot(mouse_weight, regress_values, color='green')

plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')

plt.text(20, 38, f'Correlation Coefficient: {correlation_coefficient:.2f}', color='black', fontsize=10)

plt.show()
```
